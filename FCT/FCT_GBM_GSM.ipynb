```python
# trainer and test_loaders
import torch
from torch.utils.data import DataLoader, Dataset
import pandas as pd
import numpy as np

class MyDataset(Dataset):
    def __init__(self, data, targets):
        self.data = data
        self.targets = targets

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        x = torch.tensor(self.data[idx], dtype=torch.float32)
        y = torch.tensor(self.targets[idx], dtype=torch.long)
        return x, y

batch_size = 32

train_dataset = MyDataset('/content/drive/MyDrive/Cancer data embedding/Gliosarcoma/train_embedding_200fold.pth', '/content/drive/MyDrive/Cancer data embedding/Gliosarcoma/test_embedding_200fold.pth')
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

test_dataset = MyDataset('/content/drive/MyDrive/Cancer data embedding/Gliosarcoma/train_embedding_200fold.pth', '/content/drive/MyDrive/Cancer data embedding/Gliosarcoma/test_embedding_200fold.pth')
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True)

__builtins__.test_loader = test_loader
__builtins__.train_loader = train_loader
```
